```python
import matplotlib.pyplot as plt
import seaborn as sns

# Funzione per identificare gli outlier
def find_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR)))]
    return outliers

# Seleziona le colonne numeriche per analizzare gli outlier
numerical_cols = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Trova gli outlier per ogni colonna numerica
outliers = df[numerical_cols].apply(find_outliers).dropna(how='all')

# Visualizza gli outlier su un grafico
plt.figure(figsize=(15, 10))
for col in numerical_cols:
    sns.boxplot(x=df[col])
    plt.title(f'Outliers in {col}')
    plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold

# Carica il file CSV
file_path = 'C:\\Users\\xdmat\\Desktop\\scuola\\Robotica\\Pokemon Luigi.csv'
pokemon_data = pd.read_csv(file_path)

# Dividi il dataset in features (X) e target (y)
X = pokemon_data.drop(columns=['Type 1'])  # Features
y = pokemon_data['Type 1']  # Target

# Codifica le features categoriali usando One-Hot Encoding
encoder = OneHotEncoder(sparse_output=False)
X_encoded_categorical = pd.DataFrame(encoder.fit_transform(X.select_dtypes(include=['object', 'category'])))
X_encoded_categorical.columns = encoder.get_feature_names_out(X.select_dtypes(include=['object', 'category']).columns)

# Aggiungi le colonne numeriche
X_encoded_numerical = X.select_dtypes(exclude=['object', 'category']).reset_index(drop=True)

# Concatena le colonne categoriali e numeriche
X_encoded = pd.concat([X_encoded_categorical, X_encoded_numerical], axis=1)

# Assicura che tutti i nomi delle colonne siano stringhe
X_encoded.columns = X_encoded.columns.astype(str)

# Definisci il modello di classificazione 
model = DecisionTreeClassifier()

# Definisci lo schema di cross-validation 
kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

# Esegui la cross-validation
scores = cross_val_score(model, X_encoded, y, cv=kfold, scoring='accuracy')

# Stampa le prestazioni medie della cross-validation
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
